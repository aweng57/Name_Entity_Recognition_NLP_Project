```python
import pandas as pd
import sklearn
from sklearn. preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('ner_datasetreference.csv', encoding='latin1')
df
```

```python
df.head()
```

```python
df = df.fillna(method = 'ffill')
df.head(25)
```

```python
df['Sentence #'] = LabelEncoder().fit_transform(df['Sentence #'])
df.head(25)
```

```python
df.rename(columns = {'Sentence #':'sentence_id', 'Word':'words', 'Tag':'labels'}, inplace = True)
df.head(25)
```

```python
df['labels'] = df['labels'].str.upper()
df.head(25)
```

```python
label = df['labels'].unique().tolist()
label
```

```python
sentence_ids = df['sentence_id'].unique().tolist()
train_sentence_ids, test_sentence_ids = train_test_split(sentence_ids, test_size=0.2, random_state=42)
train_data = df[df['sentence_id'].isin(train_sentence_ids)]
test_data = df[df['sentence_id'].isin(test_sentence_ids)]
```
